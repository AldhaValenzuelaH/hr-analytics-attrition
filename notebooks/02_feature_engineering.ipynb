```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'C:\Users\Aldhair\Desktop\hr-analytics-attrition\data\raw\WA_Fn-UseC_-HR-Employee-Attrition.csv', sep = ',')

print (f"Filas: {df.shape[0]}, Columnas: {df.shape[1]}")
df.head()
```

```python
#Verificamos columnas con un valor único
for col in df.columns:
    if df[col].nunique() == 1:
        print(f"{col}: valor único = {df[col].unique()}")
```

```python
columnas_a_eliminar = ['EmployeeCount', 'Over18', 'StandardHours', 'EmployeeNumber']

df = df.drop(columns=columnas_a_eliminar)

print (f"Columnas restantes: {df.shape[1]}")
df.columns.tolist()
```

```python
columnas_texto = df.select_dtypes(include='object').columns.tolist()
print(columnas_texto)

for col in columnas_texto:
    print(f"\n{col}: {df[col].unique()}")
```

```python
# Convertimos Yes/No y Female/Male a 0/1
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
df['OverTime'] = df['OverTime'].map({'Yes': 1, 'No': 0})
df['Gender'] = df['Gender'].map({'Female': 0, 'Male': 1})

df[['Attrition', 'OverTime', 'Gender']].head()
```

```python
columnas_multiples = ['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus']

df = pd.get_dummies(df, columns=columnas_multiples, drop_first=True)

print (f"Columnas ahora: {df.shape[1]}")
df.head()
```
