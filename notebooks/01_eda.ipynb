```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

#cargamos el dataset 
df = pd.read_csv(r'C:\Users\Aldhair\Desktop\hr-analytics-attrition\data\raw\WA_Fn-UseC_-HR-Employee-Attrition.csv', sep = ',')

print (f"Filas: {df.shape[0]}, Columnas: {df.shape[1]}")
df.head()
```

```python
#variable
print(df['Attrition'].value_counts())
print()
print(df['Attrition'].value_counts(normalize=True)*100)

#Gráfico
sns.countplot(data=df, x='Attrition', palette='Set2')
plt.title('Distribución de la variable Attrition (Rotación de Personal)')
plt.ylabel('Número de Empleados')
plt.show()
```

```python
#Histograma
columnas_interes=['Age', 'MonthlyIncome','DistanceFromHome','YearsAtCompany']
df[columnas_interes].hist(figsize=(10, 8),bins=20)
plt.tight_layout()
plt.show()
```

```python
columnas_interes = ['Age', 'MonthlyIncome', 'DistanceFromHome', 'YearsAtCompany']

fig, axes = plt.subplots(2,2, figsize=(12,8))
axes = axes.flatten()

for i, col in enumerate(columnas_interes):
    sns.boxplot(data=df, x='Attrition', y=col, ax=axes[i], palette='Set2')
    axes[i].set_title(f'{col} vs Attrition')

plt.tight_layout()
plt.show()
```

```python
columnas_categoricas = ['OverTime', 'Department', 'JobSatisfaction', 'WorkLifeBalance']

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, col in enumerate(columnas_categoricas):
    sns.countplot(data=df, x=col, hue='Attrition', ax=axes[i], palette='Set2')
    axes[i].set_title(f'{col} vs Attrition')

plt.tight_layout()
plt.show()
```

```python
#Solo columnas numéricas
df_numerico = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(14, 10))
sns.heatmap(df_numerico.corr(), annot=True, cmap='coolwarm', center=0)
plt.show()
```

## Interpretaciones del EDA
- El dataset muestra calidad en su contenido, ya que no muestra nulos. Sin embargo, desbalanceado en el target al indicar que solo el 16.1% presenta Attrition = Yes.
- **OverTime** es un fuerte predictor, debido a que la proporción de renuncia es notablemente mayor entre quienes hacen horas extras.
- **JobSatisfaction** tiene relación inversa con Attrition, se logra concluir que a menos satisfacción, mayor probabilidad de renuncia.
- Se encontró una relación con los empleados que renuncian, ellos tienen a ser jovenes con menor salario, mayor distancia al trabajo y con una antigüedad en la empresa menor a 5 años.
- **JobLevel** y **MonthlyIncome** están altamente correlacionadas (0.95), variable a considerar en feature engineering para evitar redundancia.
